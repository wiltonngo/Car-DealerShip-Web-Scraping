```python

from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
### HTTP Request
```

```python
# Store website in variable
```

```python
website = 'https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=mercedes_benz&models%5B%5D&list_price_max&maximum_distance=all&zip'
```

```python
## Get Request
```

```python
reponse = requests.get(website)
```

```python
## Soup Object
```

```python
soup = bs(reponse.content, 'html.parser')
```

```python
soup
```

```python
### Results
```

```python
results = soup.find_all('div',{'class':'vehicle-card'})
```

```python
### Put everything together inside a For-loop
```

```python
name= []
mileage = []
dealer_name = []
rating = []
rating_count = []
price = []

for result in results:
    try:
        name.append(result.find('h2').get_text())
    except:
        name.append('n/a')
        
    try:
        mileage.append(result.find('div',{'class':'mileage'}).get_text())
    except:
        mileage.append('n/a')
        
    try:
        dealer_name.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
    except:
        dealer_name.append('n/a')
                           
    try:
        rating.append(result.find('span',{'class':'sds-rating__count'}).get_text())
    except:
        rating.append('n/a')
                           
    try:
        rating_count.append(result.find('span',{'class':'sds-rating__link'}).get_text())
    except:
        rating_count.append('n/a')
                           
    try:
        price.append(result.find('span',{'class':'primary-price'}).get_text())
    except:
        price.append('n/a')
```

```python
## Create Pandas DataFrame
```

```python
car_dealer = pd.DataFrame({'Name':name,'Mileage': mileage, 'Dealer Name': dealer_name, 'rating':rating,
                          'Rating Count': rating_count, 'Price': price})
```

```python
car_dealer
```

```python
### Data Cleaning
```

```python
car_dealer['Rating Count'] = car_dealer['Rating Count'].apply(lambda x:x.strip('reviews').strip('('))
```

```python
car_dealer
```

```python
## Output in Excel
```

```python
car_dealer.to_excel('Single_page_car.xlsx', index = False)
```

```python
### Page2_ Pagination
```

```python
name = []
mileage = []
dealer_name = []
rating = []
rating_count = []
price = []
for i in range(1,11):
    
    website = 'https://www.cars.com/shopping/results/?page= ' + str(i) +'&page_size=20&list_price_max=&makes[]=mercedes_benz&maximum_distance=all&models[]=&stock_type=cpo&zip='
    
    response = requests.get(website)
    
    soup = bs(response.content, 'html.parser')
    
    results = soup.find_all('div',{'class':'vehicle-card'})
    
    for result in results:
        try:
            name.append(result.find('h2').get_text())
        except:
            name.append('n/a')
        
        try:
            mileage.append(result.find('div',{'class':'mileage'}).get_text())
        except:
            mileage.append('n/a')
        
        try:
            dealer_name.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
        except:
            dealer_name.append('n/a')
                               
        try:
            rating.append(result.find('span',{'class':'sds-rating__count'}).get_text())
        except:
            rating.append('n/a')
                           
        try:
            rating_count.append(result.find('span',{'class':'sds-rating__link'}).get_text())
        except:
            rating_count.append('n/a')
                               
        try:
            price.append(result.find('span',{'class':'primary-price'}).get_text())
        except:
            price.append('n/a')
```

```python
car_dealer = pd.DataFrame({'Name':name,'Mileage': mileage, 'Dealer Name': dealer_name, 'rating':rating,
                          'Rating Count': rating_count, 'Price': price})
```

```python
car_dealer
```

```python
car_dealer['Rating Count'] = car_dealer['Rating Count'].apply(lambda x:x.strip('reviews)').strip('('))
```
